```python
cat_sorted = sorted(df['stato_matrimoniale'].unique())
cat_map = {val: idx for idx, val in enumerate(cat_sorted)}

df['code_matrimoniale'] = df['stato_matrimoniale'].map(cat_map)
```

```python
df = pd.get_dummies(df, columns=['stato_matrimoniale'])
```

```python
df
```

```python
df_grouped = df.groupby(['età', 'code_matrimoniale']).agg(
    n_persone=('nomi', 'count')
).reset_index()

df_grouped
```

```python
df_pivot = df_grouped.pivot(
    index='età', 
    columns='code_matrimoniale', 
    values='n_persone'
)
df_pivot = df_pivot.fillna(0).astype(int)
df_pivot
```
