```python
from sklearn.datasets import load_diabetes
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = load_diabetes(as_frame=True)
X = ds.data          # pandas DataFrame (442, 10)
y = ds.target        # pandas Series (442,)
df = ds.frame        # DataFrame con feature + target (442, 11)

print(df.info())

scaler = StandardScaler()
```

```python
plt.scatter(df['bmi'], y)
plt.show()
plt.close()
```

```python
regr = linear_model.LinearRegression()

regr.fit(df['bmi'].values.reshape(-1, 1), y)

y_prev = regr.predict(df['bmi'].values.reshape(-1, 1))

plt.scatter(df['bmi'], y)
plt.plot(df['bmi'], y_prev, 'r')
plt.show()
plt.close()
```
